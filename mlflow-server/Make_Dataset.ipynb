```python
def choose_configuration_1(start, end):
    import random
    configuration = random.randrange(start, end)
    return configuration
```

```python
def generate_yaml(cat, xmin, ymin, xmax, ymax):
  w_img = 320 # largeur de mon image
  h_img = 320 # hauteur de mon image

  dw = 1./(w_img)
  dh = 1./(h_img)
  x = ((xmin + xmax)/2.0 - 1) / w_img
  y = ((ymin + ymax)/2.0 - 1) / h_img
  w = (xmax - xmin) / w_img
  h = (ymax - ymin) / h_img


  yaml = str(cat) + ' ' + str(x) + ' ' + str(y) + ' ' + str(w) + ' ' + str(h) + '\n'
  return (yaml)
```
